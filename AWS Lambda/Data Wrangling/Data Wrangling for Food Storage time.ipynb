```python
for i in range(0,127):
    if 'day' in df3.iloc[i,1]:
        position=df3.iloc[i,1].find('days')
        df3.iloc[i,1]=int(df3.iloc[i,1][:position-1])
    elif 'week' in df3.iloc[i,1]:
        position=df3.iloc[i,1].find('week')
        df3.iloc[i,1]=7*int(df3.iloc[i,1][:position-1])
    elif 'month' in df3.iloc[i,1]:
        position=df3.iloc[i,1].find('month')
        df3.iloc[i,1]=30*int(df3.iloc[i,1][:position-1])
    elif 'year' in df3.iloc[i,1]:
        position=df3.iloc[i,1].find('year')
        df3.iloc[i,1]=365*int(df3.iloc[i,1][:position-1])
    elif 'hour' in df3.iloc[i,1]:
        position=df3.iloc[i,1].find('hour')
        df3.iloc[i,1]=1/24*int(df3.iloc[i,1][:position-1])
```
